```python
with open("./output/combined_text.txt","r",encoding="utf-8") as f:
    text_sequence = f.read()

len(text_sequence)
```

## BPE Algorithm

```python
# ! git clone https://github.com/karpathy/minbpe.git
```

```python
from minbpe.minbpe import BasicTokenizer

tokenizer = BasicTokenizer()
tokenizer.train(text_sequence,vocab_size=1024)
```

## Visualize the vocabulary.



```python
vocab = tokenizer.vocab
vocab
```

```python
tokenizer.encode("I am sumit")
```

```python
tokenizer.decode([724, 115, 409, 304])
```

## Add special token to the vocabulary. 

### These tokens are going to be used a lot in the fine-tuning steps

```python
max_vocab_id = list(vocab.keys())[-1]
print(max_vocab_id)
tokenizer.special_tokens = {
    "<|startoftext|>": max_vocab_id + 1,
    "<|separator|>": max_vocab_id + 2,
    "<|endoftext|>": max_vocab_id + 3,
    "<|unk|>": max_vocab_id + 4,
    "<|padding|>": max_vocab_id + 5,
}
```

```python
len(tokenizer.encode(text_sequence))
```
